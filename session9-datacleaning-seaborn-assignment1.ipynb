```python
from sklearn.datasets import load_iris
import pandas as pd
iris = load_iris()
data = pd.DataFrame(iris.data,columns= iris.feature_names)
label = pd.DataFrame((list(map(lambda x: iris.target_names[x],iris.target))),columns=['species'])
iris = pd.concat([data,label],axis=1)
print(iris.head())
```

1. use the distplot() to see the distribution of SepalLengthCm,SepalWidthCm,PetalLenghtCm,PetalWidthCm features. Plot them as subplots in single image.

```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
import seaborn as sns

iris = load_iris()
# recast into long format 
iris = pd.DataFrame(data=np.c_[iris['data'], iris['target']], 
                    columns=iris['feature_names'] + ['target'])

# recast into long format 
df = iris.melt(['target'], var_name='cols',  value_name='vals')
g = sns.FacetGrid(df, col='cols', hue="target", palette="Set1")
g = (g.map(sns.distplot, "vals", hist=False, rug=True))
```

2. Do a boxplot of all features except 'Spices'

```python
sns.boxplot(data=iris.iloc[:,0:4])
```

3. Do a count plot for the feature 'Species'

```python
ax = sns.countplot( x='species',data=iris)
```

4. Do a pair plot on the features SepalLengthCm,SepalWidthCm,PetalLengthCm,PetalWidthCm,Spices

```python
sns.pairplot(iris, hue='species', size=2.5);
```

5. Do an lmplot on the following SepalLengthCm,PetalLenghtCm. Using hue, display the different spices in different colors.

```python
sns.lmplot(x="sepal length (cm)", y="petal length (cm)", hue="species", data=iris)
```

6. Do a bar plot of 'Spices' Vs 'SepalLenghtCm'

```python
ax = sns.barplot(x="sepal length (cm)", y="species", data=iris)
```

7. Using the heatmap, plot the correlation matrix

```python
# calculate the correlation matrix
corr = iris.corr()
# plot the heatmap
sns.heatmap(corr, 
        xticklabels=corr.columns,
        yticklabels=corr.columns)
```
